```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 10
```

```python
# 1. 라인 플롯 
```

```python
plt.plot([1,4,9,16])   # x = [0,1,2,3] 가정해서
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.title('y = x**2 그래프')
plt.grid()
plt.show()
```

- 스타일 지정 

```python
# 색깔, 마커, 선종류
plt.plot([1,2,3,4], [1,4,9,16], 'rs--')
plt.title("rs--' 스타일")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16], c='black', lw=3, marker='o',ms=10)
plt.title("rs--' 스타일")
plt.grid()
plt.show()
```

- 그림 범위 지정 

```python
plt.plot([1,2,3,4], [1,4,9,16], c='black', lw=3, marker='o',ms=10)
plt.title("x축, y축의 범위 설정")
plt.xlim(0,5)
plt.ylim(-2,20)
plt.grid()
plt.show()
```

- 틱 설정

```python
x = np.linspace(-np.pi, np.pi, 300)
y = np.cos(x)
plt.plot(x,y)
plt.xticks([-np.pi, -np.pi/2, 0 ,np.pi/2, np.pi])
plt.yticks([-1,-0.5,0,0.5,1])
plt.grid()
plt.show()
```

- 범례 legend

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.grid()
plt.show()
```
