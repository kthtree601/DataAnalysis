- addr2.csv를 읽어서 지도위에 표시하기 

```python
df = pd.read_csv('addr2.csv')
df
```

```python

map = folium.Map(
    location=[df['위도'].mean(), df['경도'].mean()], 
    zoom_start=13
)
for i in df.index:
    folium.Marker(
        location=[df['위도'][i], df['경도'][i]],
        tooltip=df['건물명'][i],
        popup=folium.Popup(
            df['도로명주소'][i], min_width=200, max_width=200)
    ).add_to(map) 
title_html = '<h3 align="center" style="font-size:20px"><b>6개 지역 표시</b></h3>'   
map.get_root().html.add_child(folium.Element(title_html))
map
```
