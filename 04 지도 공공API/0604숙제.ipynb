```python
df=pd.read_csv('real.park.csv')
df
```

```python
map = folium.Map(
    location=[df['X좌표(GRS80TM)'].mean(), df['Y좌표(GRS80TM)'].mean()], 
    zoom_start=13
)
for i in df.index:
    folium.Circle(
        location=[df['X좌표(GRS80TM)'][i], df['Y좌표(GRS80TM)'][i]],radius=50,
        tooltip=df['공원명'][i], 
        popup=folium.Popup(
            df['공원주소'][i], min_width=200, max_width=200),color="blue"
    ).add_to(map) 
title_html = '<h3 align="center" style="font-size:20px"><b>서울 전지역 공원 지도</b></h3>'   



map.get_root().html.add_child(folium.Element(title_html))
map
```
